```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
print(data)
data = data.cumsum()
print(data)
```

```python
data.plot()
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=list("ABCD"))
data = data.cumsum()
print(data.head(3)) # 默认是前5行数据 head(3),改为前3行
data.plot()
plt.show()
```

plot的方法：bar条形图 hist   box   kde    area   scatter   hexbin  pie等等

```python
ax = data.plot.scatter(x='A',y='B',color='Green',label='Class 1')
data.plot.scatter(x='A',y='C',color='Red',label='Class 2',ax=ax) # ax = ax表示把这张图会在上一行的ax上
plt.show()
```
